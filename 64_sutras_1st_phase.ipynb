```python
import re
import pandas as pd

with open('vedic_framework.md', 'r') as f:
    content = f.read()

# Split the content by the numeric pattern at the start of lines
raw_blocks = re.split(r'\n(?=\d+\.)|(?<=^)(?=\d+\.)', content)

concepts = []
for block in raw_blocks:
    # Extract Title: Look for number followed by text until newline or bullet
    title_match = re.search(r'^(\d+\.[^\n•]+)', block.strip())
    # Extract Application: Look for text following 'Application:'
    app_match = re.search(r'Application:\s*(.*)', block, re.DOTALL | re.IGNORECASE)

    if title_match and app_match:
        # Clean up application: take until next logical section divider
        app_text = app_match.group(1).split('\n\n')[0].split('\n•')[0].strip()
        concepts.append({
            'Concept': title_match.group(1).strip(),
            'Application': app_text.replace('\n', ' ')
        })

df_concepts = pd.DataFrame(concepts)
# Extract numeric level for proper sorting and duplicate removal
df_concepts['Level_Num'] = df_concepts['Concept'].str.extract(r'^(\d+)').astype(int)
df_concepts = df_concepts.drop_duplicates(subset=['Level_Num']).sort_values('Level_Num').drop(columns=['Level_Num']).reset_index(drop=True)

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
print(f'Total unique levels parsed: {len(df_concepts)}')
display(df_concepts)
```
